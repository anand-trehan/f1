```python
def lookup (df3, team, points):
    df3['lookup1'] = df3.season.astype(str) + df3[team] + df3['round'].astype(str)
    df3['lookup2'] = df3.season.astype(str) + df3[team] + (df3['round']-1).astype(str)
    new_df = df3.merge(df3[['lookup1', points]], how = 'left', left_on='lookup2',right_on='lookup1')
    new_df.drop(['lookup1_x', 'lookup2', 'lookup1_y'], axis = 1, inplace = True)
    new_df.rename(columns = {points+'_x': points+'_after_race', points+'_y': points}, inplace = True)
    new_df[points].fillna(0, inplace = True)
    return new_df
```
